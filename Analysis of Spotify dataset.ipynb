## Q2.How many unique singers contribute to each playlist, and which playlist has the highest diversity of singers?

```python
unique_playlist=df['playlist'].unique()
```

```python
unique_playlist
```

```python
df2_1=df[(df['playlist'] == "Today’s Top Hits")]
```

```python
df2_1
```

```python
len(df2_1['singer'].unique())
```

```python
max=-1
final_playlist=""
for playlist in unique_playlist:
    df2=df[(df['playlist'] == playlist)]
    length=len(df2['singer'].unique())
    print(f"{length} number of unique singer contribute the {playlist} ")
    if length>max:
        max=length
        final_playlist=playlist

print(f"The {final_playlist}  playlist contains songs that refer to the most unique singers, which is equal to  {length}.")
```

And Hence , Sleep playlist contain the songs of high diversity of singers

## Q3.Identify the top album/albums with the most songs in playlists, and for each of these albums, what is the total duration of their songs across all playlists?
